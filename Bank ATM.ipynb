```python
#Main Banking Program#

def banking():
    
    #Variables#
    menu_state = True
    login_state = False
    banking_stae = False
    new_old_user = 0
    with_dep = 0
    
    print("Welcome to JDLYS Banking")
    
    #Log-in || User Registration
    
    while menu_state == True:
        
        while new_old_user == 0:
            new_old_user = int(input("Existing users [1] | New Users [2]"))
            
            if new_old_user == 1:
                menu_state = False
                login_state = True
                
            elif new_old_user == 2:
                print("New user registration")
                registration()
                print("Thank you for registering!")
                menu_state = False
                
    while login_state == True:
        
        loginame = str(input("Please enter Login Name:" + "\n"))
        
        if type(globals()[loginame]) == client:
            clear_output()
            print(f'Welcome back: {globals()[loginame].fullname}')
            
            loginpin = input('Please enter your PIN')
            
            if globals()[loginame].pin != loginpin:
                print("Incorrect Pin")
            elif globals()[loginame].pin == loginpin:
                login_state = False
                banking_state = True
        
        else:
            print("User Not Found")
            
    #Banking#
    
        while banking_state == True:
            print(f'Current Balance:{globals()[loginame].bal}')
            with_dep = int(input("Withdrawals[1] | Deposits [2] | Exit [3]:"))
            
            if with_dep == 1:
                withdrawal(loginame)
                with_dep == 0
            
            elif with_dep == 2:
                deposit(loginame)
                with_dep == 0
            
            elif with_dep == 3:
                print("Thank you choosing JDLYS BANKING!")
                banking_state = False
                
            else:
                print("Pleae Choose Valid Response!")
```

```python
def registration():
    #Creates 'Client Class' with user input for Name & PIN#
    
    #Variables 
    first_name = ''
    last_name = ''
    fullname = ''
    pin = ''
    accnum = ''
    confirm = ''
    
    #Name input/confirmation
    
    while fullname == '':
        first_name = str(input("First Name:"))
        last_name = str(input("Last Name:"))
        confirm = input(f"[{first_name} {last_name}]): is this correct?[Y/N]" + "\n").title()
        
        if confirm[0] == 'N':
            continue
        if confirm[0] == 'Y':
            fullname = first_name + " " + last_name
            
        confirm = ''
        
        #PIN creation#
        while pin == '':
            
            pin = input("Please Provide a 4 digit PIN code:")
            
            if pin.isalnum == True:
                pin = ''
                print("Please provide a valid PIN code")
                continue
            elif len(pin) > 4 or len(pin) < 4:
                pin = ''
                print("Please provide a valid PIN code")
                continue
                
        #Account Number Creation#
        
        for n in range (0,6):
            accnum += str(randint(0,6))
            
        #Account to Class Allocation#
        
        account = str(fullname[0:3]) + str(accnum[0:3])
        new_user = globals()[account] = client(fullname,accnum,pin,0)
        
        print("Login Details" + "\n")
        print(f"Acc #: {account}")
        print(f"PIN:***{pin[3]}")
        
    return new_user
```

```python
def withdrawal(loginame):
    #Variable
    withdraw = 0
    
    while withdraw == 0:
        withdraw = int(input("How much would you like to withdraw?"))
        
        if (globals()[loginame].bal) < withdraw:
            withdraw = 0
            print(f"Insufficient Funds, Current Bal: ${globals()[loginame].bal}")
        
        else:
            confirm =input(f"Withdrawing ${withdraw}, is this ok? [Y/N]").title()
            
            if confirm[0] == 'N':
                withdraw = 0
            elif confirm[0] == 'Y':
                globals()[loginame].bal -= withdraw
                print(f'Current Balance: ${globals()[loginame].bal}')
                
    return globals()[loginame].bal
```

```python
def deposit(loginame):
    #Variable
    deposit = 0
    
    while deposit == 0:
        
        deposit = int(input("How much would you like to deposit?"))
        confirm =input(f"Depositing ${deposit}, is this ok? [Y/N]").title()
        
        if confirm[0] == 'N':
            deposit = 0
        if confirm[0] == 'Y':
            globals()[loginame].bal += deposit
            print(f'Current Balance: ${globals()[loginame].bal}')
    
    return globals()[loginame].bal
```

```python
class client:  
    def __init__(self,fullname,accnum,pin,bal):
        self.fullname = fullname
        self.accnum = accnum
        self.pin = pin
        self.bal = bal
    
    def __str__(self):
        return self.bal
```
